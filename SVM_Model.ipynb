#Stock Market Prediction: SVM Model Setup

```python
# IMPORTS

from sklearn.model_selection import train_test_split
from sklearn.svm import SVR

import pandas as pd
import datetime as dt
import numpy as np
```

```python
# DATAFRAME CREATION

def LoadData(fileLocation):
    df = pd.read_csv(fileLocation, delimiter = ',')
    return df
```

##Dataframe Cleaning

```python
# CLEAN DATASET

def CleanData(df):

    print("Rows before cleaning: " + str(len(df)))

    OpenIntOne = 0

    # REMOVE DUPLICATE ROWS
    df.drop_duplicates

    # REMOVE UNUSABLE DATA ROWS
    for i in (df.columns):
        for j in (range(len(df))):

            # If row has empty field
            if (pd.isnull(df.loc[j, i])):
                df = df.drop(j)

            # If value in row does not match expected type
            if (
                (i == "Date") and
                (isinstance(df.loc[j, i], str) == False)
            ):
                df = df.drop(j)

            elif (
                ((i == "Open") or (i == "High") or (i == "Low") or (i == "Close")) and
                (isinstance(df.loc[j, i], np.float64) == False)
            ):
                df = df.drop(j)

            elif (
                (i == "Volume") and
                (isinstance(df.loc[j, i], np.int64) == False)
            ):
                df = df.drop(j)

            elif (
                (i == "OpenInt") and
                (df.loc[j,i] != 0)
            ):
                OpenIntOne += 1

    # Drop last column if all same
        # May drop this if last column is necessary
    if (OpenIntOne == 0):
        df = df.drop(columns=["OpenInt"])

    print("Rows after cleaning: " + str(len(df)) + "\n")

    return df
```
